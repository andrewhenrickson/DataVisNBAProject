```python
#Extract Class
class StatisticsExtractor:
    def __init__(self, df):
        self.df = df
        self.team_home_away_list = []
        self.win_list = []
        self.opponent_list = []
        self.date_list = [] 
        self.fg2_percentage_list = []
        self.fg2_made_list = []
        self.fg2_attempted_list = []
        self.fg3_made_list = []
        self.fg3_attempted_list = []
        self.fg3_percentage_list = []
        self.free_throw_made_list = []
        self.free_throw_attempted_list = []
        self.free_throw_percentage_list = []
        self.rebound_offensive_list = []
        self.rebound_defensive_list = []
        self.rebound_total_list = []
        self.assist_list = []
        self.steal_list = []
        self.turnover_list = []
        self.blocks_list = []
        self.offensive_fouls_list = []
        self.shooting_fouls_drawn_list = []
        self.possesions_list = []
        self.defensive_fouls_list = []
        self.points_list = []





    def get_column(self, column_name):
        return self.df[column_name]

    def get_home_away(self, column):
        for i in column:
            self.team_home_away_list.append(i)

    def get_win(self, column):
        for i in column:
            self.win_list.append(i)

    def extract_def_team_and_date(self, def_team_col, date_col):
        for def_team, date in zip(def_team_col, date_col):
            self.opponent_list.append(def_team)
            self.date_list.append(date)

    def fg2_percentage(self, fg2_col, fga2_col):
        for fg, fga in zip(fg2_col, fga2_col):
            pct = fg/ fga if fga != 0 else 0
            self.fg2_percentage_list.append(pct)
    
    def fg2_made_attempted(self, fg2_col, fga2_col):
        for fg, fga in zip(fg2_col, fga2_col):
            self.fg2_made_list.append(fg)
            self.fg2_attempted_list.append(fga)
            
    def fg3_percentage(self, fg3_col, fga3_col):
        for fg, fga in zip(fg3_col, fga3_col):
            pct = fg/ fga if fga != 0 else 0
            self.fg3_percentage_list.append(pct)

    def fg3_made_attempted(self, fg3_col, fga3_col):
        for fg, fga in zip(fg3_col, fga3_col):
            self.fg3_made_list.append(fg)
            self.fg3_attempted_list.append(fga)

    def free_throw_made_attempted(self, ft_col, fta_col):
        for ft, fta in zip(ft_col, fta_col):
            self.free_throw_made_list.append(ft)
            self.free_throw_attempted_list.append(fta)
    
    def ft_percentage(self, ft_col, fta_col):
        for ft, fta in zip(ft_col, fta_col):
            pct = ft/ fta if fta != 0 else 0
            self.free_throw_percentage_list.append(pct)

    def get_rebound_number(self, rebound_col_offensive, rebound_col_defensive, rebound_col_total):
        for rebound in rebound_col_offensive:
            self.rebound_offensive_list.append(rebound)
        for rebound in rebound_col_defensive:
            self.rebound_defensive_list.append(rebound)
        for rebound in rebound_col_total:
            self.rebound_total_list.append(rebound)

    def get_assist_number(self, assist_col):
        for assist in assist_col:
            self.assist_list.append(assist)
    
    def get_steals(self, steals_col):
        for steal in steals_col:
            self.steal_list.append(steal)

    def get_turnovers(self, turnovers_col):
        for turnover in turnovers_col:
            self.turnover_list.append(turnover)
        return self.turnover_list
    
    def get_blocks(self, blocks_col):
        for block in blocks_col:
            self.blocks_list.append(block)

    def defensive_fouls(self, defensive_fouls_col):
        for foul in defensive_fouls_col:
            self.defensive_fouls_list.append(foul)

    def offensive_fouls(self, offensive_fouls_col):
        for foul in offensive_fouls_col:
            self.offensive_fouls_list.append(foul)

    def shooting_foulsdrawn(self, shooting_fouls_drawn_col):
        for foul in shooting_fouls_drawn_col:
            self.shooting_fouls_drawn_list.append(foul)
        
    def get_possesions(self, possessions_col):
        for possession in possessions_col:
            self.possesions_list.append(possession)

    def get_points(self, points_col):
        for points in points_col:
            self.points_list.append(points)

    def extract_all_statistics(self):
        self.get_home_away(self.get_column('off_home'))
        self.get_win(self.get_column('off_win'))
        self.extract_def_team_and_date(self.get_column('def_team'), self.get_column('gamedate'))
        self.fg2_percentage(self.get_column('fg2made'), self.get_column('fg2attempted'))
        self.fg2_made_attempted(self.get_column('fg2made'), self.get_column('fg2attempted'))
        self.fg3_percentage(self.get_column('fg3made'), self.get_column('fg3attempted'))
        self.fg3_made_attempted(self.get_column('fg3made'), self.get_column('fg3attempted'))
        self.ft_percentage(self.get_column('ftmade'), self.get_column('ftattempted'))
        self.free_throw_made_attempted(self.get_column('ftmade'), self.get_column('ftattempted'))
        self.get_rebound_number(self.get_column('reboffensive'), self.get_column('rebdefensive'), self.get_column('reboundchance'))
        self.get_assist_number(self.get_column('assists'))
        self.get_steals(self.get_column('stealsagainst'))
        self.get_turnovers(self.get_column('turnovers'))
        self.get_blocks(self.get_column('blocksagainst'))
        self.defensive_fouls(self.get_column('defensivefouls'))
        self.offensive_fouls(self.get_column('offensivefouls'))
        self.shooting_foulsdrawn(self.get_column('shootingfoulsdrawn'))
        self.get_possesions(self.get_column('possessions'))
        self.get_points(self.get_column('points'))

    def make_df(self):
        data = {
            'Date': self.date_list,
            'Home/Away': self.team_home_away_list,
            'Win': self.win_list,
            'Points': self.points_list,
            'Opponent': self.opponent_list,
            'FG2%': self.fg2_percentage_list,
            'FG2 Made': self.fg2_made_list,
            'FG2 Attempted': self.fg2_attempted_list,
            'FG3%': self.fg3_percentage_list,
            'FG3 Made': self.fg3_made_list,
            'FG3 Attempted': self.fg3_attempted_list,
            'FT%': self.free_throw_percentage_list,
            'FT Made': self.free_throw_made_list,
            'FT Attempted': self.free_throw_attempted_list,
            'Offensive Rebounds': self.rebound_offensive_list,
            'Defensive Rebounds': self.rebound_defensive_list,
            'Total Rebounds': self.rebound_total_list,
            'Assists': self.assist_list,
            'Steals': self.steal_list,
            'Turnovers': self.turnover_list,
            'Blocks': self.blocks_list,
            'Defensive Fouls': self.defensive_fouls_list,
            'Offensive Fouls': self.offensive_fouls_list,
            'Shooting Fouls Drawn': self.shooting_fouls_drawn_list,
            'Possessions': self.possesions_list
        }
        stats_df = pd.DataFrame(data)
        return stats_df
```

```python
allstats2023 = {}
for i in abbreviation:
    extractor2023 = StatisticsExtractor(team_dfs2023[i])
    extractor2023.extract_all_statistics()
    stats_df2023 = extractor2023.make_df()
    allstats2023[i] = stats_df2023
```

```python
def get_fg2_pct(team_stats):
    total_fg2_made = team_stats['FG2 Made'].sum()
    total_fg2_attempted = team_stats['FG2 Attempted'].sum()
    total_shooting_pct = total_fg2_made / total_fg2_attempted if total_fg2_attempted != 0 else 0
    return total_shooting_pct

def get_fg3_pct(team_stats):
    total_fg3_made = team_stats['FG3 Made'].sum()
    total_fg3_attempted = team_stats['FG3 Attempted'].sum()
    total_shooting_pct = total_fg3_made / total_fg3_attempted if total_fg3_attempted != 0 else 0
    return total_shooting_pct

def get_rebounds(team_stats):
    total_rebounds = team_stats['Total Rebounds'].sum()
    rebounds_average = total_rebounds / len(team_stats)
    return rebounds_average

def get_free_throw_pct(team_stats):
    total_ft_made = team_stats['FT Made'].sum()
    total_ft_attempted = team_stats['FT Attempted'].sum()
    total_ft_pct = total_ft_made / total_ft_attempted if total_ft_attempted != 0 else 0
    return total_ft_pct

def get_total_wins(team_stats):
    total_wins = team_stats['Win'].sum()
    return total_wins

shooting_fg2_pct_list = []
shooting_fg3_pct_list = []
ft_pct_list = []
rebounds_list = []
wins_list = []

for i in abbreviation:
    pct2 = get_fg2_pct(allstats2023[i])
    shooting_fg2_pct_list.append((pct2))
    wins = get_total_wins(allstats2023[i])
    wins_list.append((wins))
    pct3 = get_fg3_pct(allstats2023[i])
    shooting_fg3_pct_list.append((pct3))
    pct_ft = get_free_throw_pct(allstats2023[i])
    ft_pct_list.append((pct_ft))
    rebounds = get_rebounds(allstats2023[i])
    rebounds_list.append((rebounds))

```

```python
# Create a DataFrame for proper plotting
plot_data = pd.DataFrame({
    'Team': abbreviation,
    '2-Point FG%': shooting_fg2_pct_list,
    'Total Wins': wins_list,
    'Playoff Result': [playoff_results[team] for team in abbreviation]
})


# Create scatter plot
fig = px.scatter(
    plot_data,
    x='2-Point FG%',
    y='Total Wins',
    text='Team',
    color='Playoff Result',
    color_discrete_map=color_map,
    category_orders={'Playoff Result': ['Champion', 'Finals', 'Conference Finals', 
                                        'Conference Semifinals', 'First Round', 
                                        'Did Not Make Playoffs']},
    title='NBA Team Wins vs 2-Point Shooting Percentage (2023 Season)',
    hover_data={'2-Point FG%': ':.3f', 'Team': True}
)

# Customize markers and text
fig.update_traces(
    textposition='top center',
    marker=dict(size=10, line=dict(width=1, color='darkblue'))
)

for trace in fig.data:
    if trace.name == 'Champion':
        trace.marker.symbol = 'star'
        trace.marker.size = 18  
    elif trace.name == 'Finals':
        trace.marker.symbol = 'diamond'
        trace.marker.size = 15
    else:
        trace.marker.symbol = 'circle'

# Update layout
fig.update_layout(
    xaxis_title='2-Point Field Goal Percentage',
    yaxis_title='Total Wins',
    hovermode='closest',
    width=900,
    height=600,
    legend_title='Playoff Result'
)

# Format x-axis as percentage
fig.update_xaxes(tickformat='.1%')

fig.show()
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
import plotly.graph_objects as go

# Create plot_data FIRST (before the Dash app)
plot_data = pd.DataFrame({
    'Team': abbreviation,
    '2-Point FG%': shooting_fg2_pct_list,
    'Total Wins': wins_list
})

# Create scatter plot figure
scatter_fig = px.scatter(
    plot_data,
    x='2-Point FG%',
    y='Total Wins',
    text='Team',
    title='NBA Team Wins vs 2-Point Shooting Percentage (2023 Season)',
    hover_data={'2-Point FG%': ':.3f', 'Team': True}
)
scatter_fig.update_traces(textposition='top center', marker=dict(size=10))
scatter_fig.update_xaxes(tickformat='.1%')

# Create Dash app
app = Dash(__name__)

app.layout = html.Div([
    dcc.Graph(id='scatter-plot', figure=scatter_fig),
    dcc.Graph(id='detail-plot')
])

# Only need one callback for the detail plot
@app.callback(
    Output('detail-plot', 'figure'),
    Input('scatter-plot', 'clickData')
)
def display_detail(clickData):
    if clickData is None:
        return go.Figure()
    
    team_abbr = clickData['points'][0]['text']
    team_stats = allstats2023[team_abbr]
    
    fig = px.line(
        team_stats,
        y='Points',
        title=f'{team_abbr} - Points Throughout Season',
        labels={'index': 'Game Number', 'value': 'Points'}
    )
    
    return fig

if __name__ == '__main__':
    app.run(debug=True, port=8050)
```
